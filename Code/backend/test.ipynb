```python
from ipycanvas import Canvas
from ipywidgets import VBox, Button
from IPython.display import display
import numpy as np
import cv2
import torch

# Create canvas with sync_image_data=True
canvas = Canvas(width=280, height=280, sync_image_data=True)
canvas.stroke_style = "white"
canvas.line_width = 18
canvas.fill_style = "black"
canvas.fill_rect(0, 0, 280, 280)

drawing = {"active": False}

# Mouse events
def handle_mouse_down(x, y):
    drawing["active"] = True
    canvas.begin_path()
    canvas.move_to(x, y)

def handle_mouse_move(x, y):
    if drawing["active"]:
        canvas.line_to(x, y)
        canvas.stroke()

def handle_mouse_up(x, y):
    drawing["active"] = False

canvas.on_mouse_down(handle_mouse_down)
canvas.on_mouse_move(handle_mouse_move)
canvas.on_mouse_up(handle_mouse_up)

# Clear button
def clear_canvas(b):
    canvas.clear()
    canvas.fill_style = "black"
    canvas.fill_rect(0, 0, 280, 280)

clear_btn = Button(description="Clear")
clear_btn.on_click(clear_canvas)

display(VBox([canvas, clear_btn]))
```

```python
def canvas_to_tensor(canvas):
    img = np.array(canvas.get_image_data())[:, :, 0]
    img = cv2.resize(img, (112, 112))
    img = img / 255.0
    img = torch.tensor(img).unsqueeze(0).unsqueeze(0)
    return img.float().to(DEVICE)
```

```python
def predict():
    x = canvas_to_tensor(canvas)

    with torch.no_grad():
        logits = model(x)
        probs = torch.softmax(logits, dim=1)[0]

    top5 = torch.topk(probs, 5)

    print("ðŸŽ¯ Top-5 predictions:")
    for i in range(5):
        idx = top5.indices[i].item()
        print(
            f"{i+1}. {CLASSES[idx]:20s}  {top5.values[i].item():.3f}"
        )

predict()
```
